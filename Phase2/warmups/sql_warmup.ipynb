### Simple Query

Which country has the most airports? Which one has the least?

```python
pd.read_sql("""

SELECT *,COUNT()
FROM airports
GROUP BY country
ORDER BY COUNT() DESC
LIMIT 1
""", con)
```

### Where?

Which airports are in the `Pacific/Port_Moresby` time zone? What city and country?

```python
pd.read_sql("""

SELECT city,country
FROM airports
WHERE timezone = 'Pacific/Port_Moresby'
""", con)
```

### Further down the rabbit hole!

Which countries have between 50 and 100 airports?

```python
pd.read_sql("""

SELECT *,COUNT() AS count
FROM airports

GROUP BY country
HAVING count >= 50 AND count <= 100

ORDER BY count

""", con)
```

```python
pd.read_sql("""
SELECT 
    *, COUNT(routes.airline_id) as num_routes
FROM 
    airlines
JOIN 
    routes 
    ON airlines.id = routes.airline_id

""", con)
```

### Subqueries/CTEs!

What is the average number of routes per airline?

```python
pd.read_sql("""
SELECT 
    airline_id,count()
FROM 
    routes
GROUP BY 
    airline_id
""", con)
```

```python
# Get the average via a subquery
pd.read_sql("""
SELECT
    AVG(num_of_routes)
FROM 
    (
    SELECT 
        airline_id,
        count() AS num_of_routes
    FROM 
        routes
    GROUP BY 
        airline_id
    )
    
""",con)
```

```python
## Greg and Marshall
pd.read_sql("""
WITH avg_routes AS (
    SELECT
        airline, airline_id, COUNT() as num_routes
    FROM
        routes
    GROUP BY
        airline
)
SELECT
    AVG(num_routes) AS avg_num_routes
FROM
    avg_routes

""", con)
```

```python
pd.read_sql("""
SELECT COUNT(*), airline
FROM airlines
JOIN routes
    ON airlines.id = routes.airline_id
JOIN airports
    ON routes.source_id = airports.id
WHERE airports.timezone = "Pacific/Port_Moresby"
GROUP BY airline
ORDER BY COUNT(*) DESC
LIMIT 1
;""", con)
```

```python
qry = """

SELECT
    COUNT(rt.airline_id) AS rt_count,
    al.name as Name
FROM
    routes AS rt
    INNER JOIN airports AS ap ON rt.source_id = ap.id
    INNER JOIN airlines AS al on rt.airline_id = al.id
WHERE
    ap.timezone = 'Pacific/Port_Moresby'
GROUP BY
    rt.airline_id
ORDER BY
    rt_count DESC
LIMIT 1

"""

pd.read_sql(qry, con)
```
